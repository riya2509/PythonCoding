```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='reg')
```

```python
sns.pairplot(iris,hue='petal_width')
```

```python
sns.distplot(iris['petal_width'])
```

```python
sns.violinplot(x='sepal_length',y='species',data=iris,orient='h')
```
