# Random Forest

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Dataset\\Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting the Random Forest Regression Model to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10,random_state=0)#n_estimators = no. of trees
regressor.fit(X,y)
```

```python
# Predicting a new result
y_pred = regressor.predict([[6.5]])
y_pred
```

```python
# Visualising the Regression results (for higher resolution and smoother curve)
plt.figure(figsize=(15, 12))
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

We got more steps in the above stair like structure because of several descision tree instead of one decision tree.Therefore we have a lot more of splits of the whole range of levels and therefore a lot more intervals of different levels.

Each straight horizontal line here separated by these vertical lines is one interval or one split.

If we talk about prediction for 6.5 level then the thing which happened with this prediction is that ten(here) are voting on which step the salary of the 6.5 level position would be and then the random forest took the average of all the different predictionsof the salary of the 6.5 level made by all the diffrent tree in the forest.

Why adding more trees does not add more steps?


This is so because the more we add trees the more the average of the predictions made by the tree is converging to the same average.This is base based on the technique`Entropy` and `Information gain`
